```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
data= df.drop(["koi_disposition","koi_period_err2"], axis=1)
selected_features = data.columns
data.head()
```

```python
# Assign the data to X
X= pd.get_dummies(data)
X.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign the data to y
# Use Pandas get_dummies to convert categorical data
y = df["koi_disposition"]
y.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder,MinMaxScaler
from sklearn.linear_model import LogisticRegression
from tensorflow.keras.utils import to_categorical
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=42)
X_train.head()
```

```python
# Scale your data
X_scaler=MinMaxScaler().fit(X_train)
# X_scaler = StandardScaler().fit(X_train)
# y_scaler= StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Label-encode data set
label_encoder= LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
encoded_y_train
```

# Train the Model



```python
classifier = LogisticRegression()
classifier.fit(X_train_scaled, encoded_y_train)
```

```python
print(classifier.score(X_train_scaled,encoded_y_train))

print(classifier.score(X_test_scaled,encoded_y_test))
```

```python
# print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
# print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
```

```python
# Train the model with GridSearch
parameters={"penalty":['l1', 'l2', 'elasticnet', 'none'],"tol":[1e-4,1e-6,1e-2],"C":[1000,100, 0.001]}
clf= GridSearchCV(classifier,parameters)
clf.fit(X_train_scaled,encoded_y_train)
```

```python
# print(grid2.best_params_)
# print(grid2.best_score_)
print(clf.best_params_)
print(clf.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = clf.predict(X_test_scaled)
```

```python
#Classification Report 
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions,
                            target_names=["Confirmed", "False Positive","Candidate"]))
```
